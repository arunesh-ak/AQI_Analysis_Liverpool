```python
def changeDate(end_date):
    '''This function takes a date as input, increases the value of the date by 1 month, and returns it.'''
    #splitting the date into year, month and day
    end_year, end_month, end_day = map(int, end_date.split('-'))
    #if the month is 12 then it will increase 1 year and change the month to 1
    if end_month == 12:
        end_month = 1
        end_year= end_year + 1
    #this will increase the month by 1
    else:
        end_month = end_month + 1
        #if the month is less than 10 then it will append 0 at the starting od the number like 1 will become '01'
        if(end_month < 10):
            end_month = f'0{end_month}'
    #returning the date in string in proper format like '2022-01-01'
    return f'{end_year}-{end_month}-{end_day}'
```

```python
#importing requests, pandas and time
import requests
import pandas as pd
import time


def get_historical_aqi(api_key, lat, lon, start_date, end_date,i):
    '''Fetch the data from the API from start_date to end_date and return the data in json format'''
    url = f"https://api.weatherbit.io/v2.0/history/airquality?lat={lat}&lon={lon}&start_date={start_date}&end_date={end_date}&key={api_key}"
    

    response = requests.get(url)
    
    #if the data is successfully fetched then retuening the data else returning the response status code
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print("Error:", response.status_code)
        print(i)
        return None


# Replace the API Key  with your actual Weatherbit API key
api_key = '6e2d0b819549423eb6fc41ec72111840'
latitude = -33.920921  # (Australia-Liverpool)
longitude = 150.923141  # (Australia-Liverpool)
start_date = '2022-02-01'
end_date = '2022-03-01'

#calling the get_historical_aqi function and storing the setch data in historical_aqi_data
historical_aqi_data = get_historical_aqi(api_key, latitude, longitude, start_date, end_date,0)

#if the data is fetched successfully then storing the data in df a dataframe
if historical_aqi_data != None:
    df = pd.DataFrame(historical_aqi_data['data'])

#pausing the code for 5 second
time.sleep(5)

#Fetching the data 23 time more i.e for more 23 months
for i in range(1,24):
    start_date = end_date
    #Increase the date of end_date by 1 month using changeDate function
    end_date = changeDate(end_date)
    #calling the get_historical_aqi function and storing the setch data in historical_aqi_data
    historical_aqi_data = get_historical_aqi(api_key, latitude, longitude, start_date, end_date,i)
    #if the data is fetched successfully then storing the data in df2 a dataframe
    if historical_aqi_data == None:
        continue
    df2 = pd.DataFrame(historical_aqi_data['data'])

    #Concatenating df and df2 using common columns
    df = pd.concat([df, df2], ignore_index=True)
    #pausing the code for  seconds
    time.sleep(5)
    
#Print the dataframe if the data is sucessfully fetched otherwise printing error message
if historical_aqi_data:
    print(df)
else:
    print("Failed to retrieve historical AQI data.")

#displaying df
df
```
